## Duplicate Check

The dataset was checked for duplicate records, and no duplicate rows were found.

```python
df[df["Age"] < 0]
```

```python
df[df["Age"] < 0].shape
```

```python
df = df[df["Age"] >= 0]
```

```python
df[df["Age"] < 0]
```

## Handling Invalid Values

The **Age** column was checked for invalid values. One record with a negative age (-1) was identified. Since a negative age is not valid, the record was removed from the dataset to improve data quality.

```python
df[df["Age"] < 0].shape
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=20, kde=True)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Patients")

plt.show()
```

```python
plt.savefig("screenshots/age_distribution.png", bbox_inches="tight")
```

```python
plt.figure(figsize=(6,5))
sns.countplot(data=df, x="Gender")

plt.title("Gender Distribution")

plt.show()
```

```python
plt.savefig("screenshots/gender_distribution.png", bbox_inches="tight")
```

```python
plt.figure(figsize=(6,5))
sns.countplot(data=df, x="No-show")

plt.title("Appointment No-show Distribution")

plt.show()
```

```python
plt.savefig("screenshots/no_show_distribution.png", bbox_inches="tight")
```
